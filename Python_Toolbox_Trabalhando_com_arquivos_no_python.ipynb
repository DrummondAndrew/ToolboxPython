#### Carregando Planilhas

1- Carregando Panilhas do Gdrive

1.1. PLanilhas em excel (.xlxs) e CSV (.csv)

```python
import pandas as pd
from google.colab import drive

# 1. Monta o seu Google Drive no ambiente do Colab
drive.mount('/content/drive')

# 2. Defina o caminho do arquivo (substitua pelo seu caminho real)
# Geralmente os seus arquivos ficam dentro de 'My Drive' ou 'Compartilhados com comigo'
caminho_arquivo = '/content/drive/MyDrive/seu_arquivo.xlsx'

# 3. Carrega o arquivo usando o pandas
# Se for uma planilha do Excel (.xlsx)

df = pd.read_excel(caminho_arquivo)

# Se for um arquivo CSV (.csv)
df = pd.read_csv(caminho_arquivo)

# Exibe as primeiras linhas
print(df.head())
```
1.2. Selecionando abas na planilha

 - caso pretenda importar apenas uma das abas da planhilha:

```python
# defina o nome da aba a ser importada

Aba_Alvo= 'Nome da Aba'

df = pd.read_excel(caminho_arquivo, sheet_name=Aba_Alvo)
```
1.2. Carregando planilhas do tipo Google sheets
```python
import pandas as pd
from google.colab import auth
import gspread
from google.auth import default

# 1. Autentica o seu usuário
auth.authenticate_user()

# 2. Obtém as credenciais do ambiente
creds, _ = default()
gc = gspread.authorize(creds)

# 3. Abre a planilha pelo nome exato ou pela URL
# Substitua pelo nome da sua planilha no Drive
planilha = gc.open('Nome da Sua Planilha no Google Sheets')

# 4. Seleciona a primeira aba (página) da planilha
aba = planilha.sheet1

# 5. Converte os dados da aba para um DataFrame do pandas
dados = aba.get_all_records()
df = pd.DataFrame(dados)

# Exibe o resultado
print(df.head())
```
