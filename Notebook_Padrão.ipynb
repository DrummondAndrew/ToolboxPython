```python
#Teste de Funções Polinomiais
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo (0.1 para evitar log(0))
x = np.linspace(0.1, 10, 400) # Usando escala menor para visualização clara

# Definição das funções
y1 = x**2 + 3*x - 5
y2 = np.sin(x)
y3 = np.tan(x)
y4 = np.log(x) # ln(x) em numpy é np.log

# Criando a figura
plt.figure(figsize=(12, 8))

# Subplot 1: Quadrática
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.title('$y = x^2 + 3x - 5$')
plt.grid(True)

# Subplot 2: Seno
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'b')
plt.title('$y = \sin(x)$')
plt.grid(True)

# Subplot 3: Tangente (Limpando as assíntotas para melhor visão)
y3_clean = np.ma.masked_inside(y3, -10, 10) # limita a escala vertical
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'g')
plt.ylim(-10, 10)
plt.title('$y = \tan(x)$')
plt.grid(True)

# Subplot 4: Logaritmo (ln)
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'purple')
plt.title('$y = \ln(x)$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Função Exponencial
x_exp = np.linspace(0, 10, 100)
y5 = 2**x_exp + 5

plt.figure(figsize=(6, 4))
plt.plot(x_exp, y5, color='orange', label='$y = 2^x + 5$')
plt.title('Função Exponencial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gerando um Histograma
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Biblioteca excelente para estatística

# Gerando 1000 pontos de dados aleatórios
dados = np.random.normal(loc=50, scale=10, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(dados, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma dos Dados')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Gerando uma curva normal
import scipy.stats as stats

plt.figure(figsize=(8, 5))
# Criando a linha da curva normal teórica
x = np.linspace(20, 80, 100)
plt.plot(x, stats.norm.pdf(x, 50, 10), color='red', lw=3)

plt.title('Curva Normal (Densidade)')
plt.fill_between(x, stats.norm.pdf(x, 50, 10), color='red', alpha=0.2)
plt.grid(True)
plt.show()
```

```python
#Gerando um boxplot
plt.figure(figsize=(8, 5))
plt.boxplot(dados, vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightgreen'))

plt.title('Boxplot - Identificação de Quartis e Outliers')
plt.xlabel('Valores')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
